```python
import pandas as pd
import numpy as np
import tkinter as tk
from tkinter import *
import time
```

```python
# Создаем окно
game = Tk()
game.geometry("300x250")
# game.iconbitmap(r'pngegg.ico')
game.title("Игра 'Угадай число'").center(200)
sum_oper = 0
diction = ['Больше!', 'Меньше!']


def fun_game(lan):
    # Создаем фреймы для размещения виджетов
    frame_game = tk.LabelFrame(
        game, padx=20, pady=20, width=200, height=500, bd=0)
    frame_game.pack(padx=10, pady=10, fill='both')

    lan.destroy()
    left_border = 0
    right_border = 100

    label_window = tk.Label(frame_game,
                            text="Твое число:", font=('Comic Sans MS', 12)).pack()

    label_value = tk.Label(frame_game, text="", font=('Comic Sans MS', 12))
    label_value.pack()

    button_answer_1 = tk.Button(frame_game, text='Угадал!', width=12, height=2, fg="green", font=(
        'Comic Sans MS', 10)).pack()

    for i in range(0, len(diction)):
        button_answers = tk.Button(frame_game, text=diction[i], width=12, height=2, fg="green", font=(
            'Comic Sans MS', 10), command=lambda: range_param(text.get())).pack(side='left')


    def range_param(text):
        print(text)

    def set_label():
        global sum_oper
        sum_oper += 1
        currentTime = np.random.randint(1, 100)
        label_value['text'] = currentTime
        if sum_oper < 18:
            game.after(50, set_label)
        else:
            set_values()

    set_label()

    def set_values():
        value = round(right_border/2)
        label_value.config(text=str(value))
        label_value.pack(fill='both')


def click_scrin():
    # Создаем фреймы для размещения виджетов
    frame_scrin = tk.LabelFrame(
        game, padx=20, pady=20, width=200, height=500, bd=0)
    frame_scrin.pack(padx=10, pady=10, fill='both')

    label_scrin = Label(frame_scrin, text="Привет!\nЗадумай число от 1 до 100.\nЯ попытаюсь отгадать его!\n",
                        fg="#274D7E", font=('Comic Sans MS', 11), wraplength=250).pack()

    button_scrin = tk.Button(frame_scrin, text='◃ГОТОВО▹', width=8, height=2, fg="#274D7E", font=(
        'Comic Sans MS', 15), command=lambda: fun_game(frame_scrin)).pack(side='bottom')


click_scrin()
game.mainloop()
```
